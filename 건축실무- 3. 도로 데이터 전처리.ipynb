```python
import shapefile  #the pyshp module : Should install pyshp module.
import pandas as pd
from pyproj import Proj, transform  #Should install pyproj module.
```

```python
# read data (Copy all files from nodelink into nodelink folder: I made it.)
# using old_data
shp_path_node = '/content/drive/MyDrive/Colab Notebooks/건축실무/도로/2014/MOCT_NODE.shp'
sf_node = shapefile.Reader(shp_path_node, encoding= 'iso-8859-1')
shp_path_link = '/content/drive/MyDrive/Colab Notebooks/건축실무/도로/2014/MOCT_LINK.shp'
sf_link = shapefile.Reader(shp_path_link, encoding= 'iso-8859-1')

# construct pandas dataframe
#grab the shapefile's field names
# node
fields_node = [x[0] for x in sf_node.fields][1:]
records_node = sf_node.records()
shps = [s.points for s in sf_node.shapes()] # node has coordinate data.
# link
fields_link = [x[0] for x in sf_link.fields][1:]
records_link = sf_link.records()

#write the records into a dataframe
#node
node_dataframe = pd.DataFrame(columns=fields_node, data=records_node)
#add the coordinate data to a column called "coords"
node_dataframe = node_dataframe.assign(coords=shps)
# link
link_dataframe = pd.DataFrame(columns=fields_link, data=records_link)
```

```python
node_dataframe
```

```python
# Data restriction
boundary=['270','271','272','273','274','275','276'] # STNL_REG for Incheon
df_node = pd.DataFrame()
df_link = pd.DataFrame()
for strs in boundary:
    res_node = node_dataframe[node_dataframe['STNL_REG'] == strs ] # STNL_REG is not int.
    res_link = link_dataframe[link_dataframe['STNL_REG'] == strs ]
    df_node = pd.concat([df_node,res_node],ignore_index=True) # marge data
    df_link = pd.concat([df_link,res_link],ignore_index=True)
```

```python
# Data restriction for nodes
df_node = pd.DataFrame()

boolean=[]
for i in range(len(node_dataframe)):
  if node_dataframe['NODE_ID'][i][:3] in ['270','271','272','273','274','275','276']:
    boolean.append(True)
  else:
    boolean.append(False)

res_node= node_dataframe[boolean]
df_node = pd.concat([df_node,res_node],ignore_index=True)
df_node
```

```python
# Data restriction for links
df_link = pd.DataFrame()

boolean=[]
for i in range(len(link_dataframe)):
  if link_dataframe['LINK_ID'][i][:3] in ['270','271','272','273','274','275','276']:
    boolean.append(True)
  else:
    boolean.append(False)

res_link= link_dataframe[boolean]
df_link = pd.concat([df_link,res_link],ignore_index=True)
df_link
```

```python
# Change coordinate system
# korea 2000/central belt 2010 (epsg:5186) to wgs84(epsg:4326)
inProj = Proj(init = 'epsg:5186')
outProj= Proj(init = 'epsg:4326')
latitude = []
longitude= []
for idx,row in df_node.iterrows():
    x,y  = row.coords[0][0],row.coords[0][1]  # korea 2000 좌표계
    nx,ny = transform(inProj,outProj,x,y)     # 새로운 좌표계    
    latitude.append(ny)
    longitude.append(nx)
df_node['latitude'] = latitude
df_node['longitude']= longitude
del df_node['coords'] # delete coords
```

```python
# Change column name to draw network in Gephi
df_node.rename(columns={'NODE_ID':'Id'},inplace = True)
df_link.rename(columns={'F_NODE':'Source','T_NODE':'Target'},inplace = True)
```

```python
df_node.to_csv('/content/drive/MyDrive/Colab Notebooks/건축실무/도로 데이터/2008_node.csv') # node list
df_link.to_csv('/content/drive/MyDrive/Colab Notebooks/건축실무/도로 데이터/2008_link.csv') # edge(=link) list
```

```python
nodes = nodes[['Id','NODE_NAME','latitude','longitude']]
links = links[['Source','Target']]

source_in = links['Source'].apply(lambda x : x in list(nodes['Id'])) # check Sources are in 분당
target_in = links['Target'].apply(lambda x : x in list(nodes['Id'])) # check Targets are in 분당
# source_in and target_in are boolean type pandas.Series which contains True or False
final_links = links[source_in & target_in] # contain if both target and source are contained in 분당
```

```python
#그래프 생성
import os
import pandas as pd
import numpy as np
import networkx as nx

G = nx.Graph()
R = 6371e3

for idx,row in nodes.iterrows():
    # add node to Graph G
    G.add_node(row['Id'],Label=row['NODE_NAME'],latitude=row['latitude'], longitude=row['longitude'])

for idx,row in final_links.iterrows():
    ## Calculate the distance between Source and Target Nodes
    '''   
    lon1 = float(nodes[nodes['Id'] == row['Source']]['longitude'] * np.pi/180)
    lat1 = float(nodes[nodes['Id'] == row['Source']]['latitude'] * np.pi/180)
    lon2 = float(nodes[nodes['Id'] == row['Target']]['longitude'] * np.pi/180)
    lat2 = float(nodes[nodes['Id'] == row['Target']]['latitude'] * np.pi/180)
    d_lat = lat2 - lat1
    d_lon = lon2 - lon1
    a = np.sin(d_lat/2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(d_lon/2) ** 2
    c = 2 * np.arctan2(a**0.5, (1-a) ** 0.5)
    d = R * c
    '''    
    # Link attribute : 'Source', 'Target' and weight = 'Length between them'
    G.add_edge(row['Source'],row['Target'],weight = 1000)


matrix=[]
for ix, row in nodes.iterrows():
  radius=G.degree[row['Id']]
  matrix.append(radius)

nodes['radius']=matrix

nodes= nodes[['Id','radius']]
```

```python
##########통째로
filedir= '/content/drive/MyDrive/Colab Notebooks/건축실무/도로 데이터'
for year in range(2021,2007,-1):
  nodes = pd.read_csv(os.path.join(filedir, str(year) +"_node.csv"))
  links = pd.read_csv(os.path.join(filedir, str(year) +"_link.csv"))

  nodes = nodes[['Id','NODE_NAME','latitude','longitude']]
  links = links[['Source','Target']]

  source_in = links['Source'].apply(lambda x : x in list(nodes['Id'])) # check Sources are in 분당
  target_in = links['Target'].apply(lambda x : x in list(nodes['Id'])) # check Targets are in 분당
  # source_in and target_in are boolean type pandas.Series which contains True or False
  final_links = links[source_in & target_in] # contain if both target and source are contained in 분당

  G = nx.Graph()
  R = 6371e3

  for idx,row in nodes.iterrows():
    # add node to Graph G
    G.add_node(row['Id'],Label=row['NODE_NAME'],latitude=row['latitude'], longitude=row['longitude'])

  for idx,row in final_links.iterrows():
    # Link attribute : 'Source', 'Target' and weight = 'Length between them'
    G.add_edge(row['Source'],row['Target'],weight = 1000)

  matrix=[]
  for ix, row in nodes.iterrows():
    radius=G.degree[row['Id']]
    matrix.append(radius)

  nodes['radius']=matrix
  nodes= nodes[['Id','radius']]
  nodes= nodes.drop_duplicates(['Id'])
  print(len(nodes))
  
  if year==2021:
    df1= nodes
  elif year<=2020:
    df1= pd.merge(df1, nodes, how='left', left_on='Id', right_on='Id')

df1= df1.fillna(0)
```
