```python
import numpy as np
import pandas as pd
```

```python
file_dir='/content/drive/MyDrive/Colab Notebooks/건축실무/인구/인구 2021.csv'

df= pd.read_csv(file_dir, delimiter= ',')
df= df.drop(index=0, axis=0)
df= df.reset_index(drop= True)

df= df[['행정구역','총인구수']]
print(len(df))

df= df[df['총인구수']!=0]
df= df[df['총인구수']!='-']
df= df.drop_duplicates(['행정구역'])
df= df.reset_index(drop= True)
print(len(df))

df= df.astype({'행정구역':'string','총인구수':'int'})
df.dtypes

boolean=[]
for i in range(len(df)):
  if df['행정구역'][i] in ['전국','서울특별시','부산광역시','대구광역시','인천광역시','광주광역시','대전광역시','울산광역시','경기도','강원도','충청북도','충청남도','전라북도','전라남도','경상북도','경상남도','제주특별자치도','세종특별자치시'] or df['행정구역'][i][-1]=='소':
    boolean.append(False)
  else:
    boolean.append(True)

df= df[boolean]

df.to_csv(file_dir, index= False, sep=',')
```
